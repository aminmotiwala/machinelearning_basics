```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder


df = pd.read_csv('sample_data/carprices.csv')
df

```

```python
le = LabelEncoder()
df['Car Model'] = le.fit_transform(df['Car Model'])
df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

dfX = df[['Car Model','Mileage', 'Age(yrs)']].values
dfX

dfY = df[['Sell Price($)']].values
dfY

ct = ColumnTransformer([('Car Model',OneHotEncoder(),[0])], remainder='passthrough')
dfXTransformed = ct.fit_transform(dfX)
dfXTransformed
```

```python
#removing first column for dummy value error
dfXTransformed = dfXTransformed[:,1:]
dfXTransformed
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(dfXTransformed,dfY)
```

```python
model.predict([[0,1,45000, 4]]) # 1) Predict price of a mercedez benz that is 4 yr old with mileage 45000
```

```python
model.predict([[1,0,86000, 7]]) # Predict price of a BMW X5 that is 7 yr old with mileage 86000
```

```python
# Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())
model.score(dfXTransformed,dfY)
```
