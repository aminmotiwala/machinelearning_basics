```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder


df = pd.read_csv('sample_data/homeprices_categorical_variable.csv')
df

```

```python
le = LabelEncoder()
df.town = le.fit_transform(df.town)
df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

dfX = df[['town','area']].values
dfX

dfY = df[['price']].values
dfY

ct = ColumnTransformer([('town',OneHotEncoder(),[0])], remainder='passthrough')
dfXTransformed = ct.fit_transform(dfX)
dfXTransformed
```

```python
#removing first column for dummy value error
dfXTransformed = dfXTransformed[:,1:]
dfXTransformed
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(dfXTransformed,dfY)
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
```

```python
model.predict([[1,0,2800]]) # 2800 sqr ft home in robbinsville
```
